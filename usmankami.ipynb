```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error
import imageio
import os
from statsmodels.graphics.tsaplots import plot_acf
```

```python
energy_hourly = pd.read_csv('C:\\Users\\usman\\Desktop\\major.csv', 
                            index_col=[0], parse_dates=[0])

#Indices are not sorted - order the readings
energy_hourly.sort_index(inplace=True)

#PJME_MW - MW per hour in PJM East Area
energy_hourly.head(3)
```

```python
def split_data(data, split_date):
    return data[data.index <= split_date].copy(), \
           data[data.index >  split_date].copy()
```

```python
train, test = split_data(energy_hourly, '1-Jul-13')

plt.figure(figsize=(15,5))
plt.xlabel('year')
plt.ylabel('blood group population')
plt.plot(train.index,train)
plt.plot(test.index,test)
plt.show()
```

```python
train, test = split_data(energy_hourly, '1-Jul-13')

plt.figure(figsize=(15,5))
plt.xlabel('year')
plt.ylabel('blood count')
plt.plot(train.index,train)
plt.plot(test.index,test)
```

```python
def create_features(df):
    
    df['date'] = df.index
    df['hour'] = df['date'].dt.hour
    df['dayofweek'] = df['date'].dt.dayofweek
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['dayofmonth'] = df['date'].dt.day
    df['weekofyear'] = df['date'].dt.weekofyear
    
    X = df[['hour','dayofweek','quarter','month','year',
           'dayofyear','dayofmonth','weekofyear']]
    return X
```

```python
X_train, y_train = create_features(train), train['Donation_Date']
X_test, y_test   = create_features(test), test['Donation_Date']


```

```python
reg = xgb.XGBRegressor(n_estimators=1000)
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        early_stopping_rounds=50, #stop if 50 consequent rounds without decrease of error
        verbose=False)
```

```python
xgb.plot_importance(reg, height=0.9)
```
